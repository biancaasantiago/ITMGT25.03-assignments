```python
import matplotlib.pyplot as plt
import requests
import time
import json
```

```python
data_points = []

for i in range (1,126,1):
    data = requests.get('http://api.open-notify.org/iss-now.json')
    iss_data = data.json()
    latitude = iss_data['iss_position']['latitude']
    longitude = iss_data['iss_position']['longitude']
    timestamp = iss_data['timestamp']
    data_points.append((latitude, longitude, timestamp))
    time.sleep(10)
    
print(data_points)
```

```python
lat_data = [lat_point[0] for lat_point in data_points]
long_data = [long_point[1] for long_point in data_points]

plt.figure(figsize=(10, 6))
plt.plot(lat_data, long_data, color='pink')
plt.title('Path of the International Space Station')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

plt.show()
```
